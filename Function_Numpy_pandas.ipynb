   ```python
def function_name(_Input_):
    do_something with _Input_ to produce _Output_
    return _Output_
```

    function มีส่วนสำคัญทั้งหมด 4ส่วน
        1. บอก python ว่าเราจะเขียนฟังก์ชัน ชื่ออะไร def function_name():(ขาดไม่ได้)
        2. กำหนดตัวแปรที่จะเป็น input(ขาดได้)
        3. ส่วนประมวลผล do_something with Input to get Output (ขาดไม่ได้)
        4. ส่วน output return Output(ขาดได้)

## HW3 ให้เขียน function คำนวณวินาทีจาก เวลาต่อไปนี้
   ## '14-15-31'
   ## '13-00-01'
   ## '05-49-44'
   ## '00-00-16'
   ## เเละ print ให้สวยงาม

````
def time(Heart):
  x1=  int (Heart.split('-')[0])*3600
  y1= int (Heart.split('-')[1])*60
  z1 = int (Heart.split('-')[2])
  t = x1+y1+z1
  print("แปลงเป็นวินาทีได้ = ",t," วินาที")
````

````
time('14-15-31') 
````

````
time('13-00-01')
````

````
time('05-49-44')
````

````
time('00-00-16')
````

## อินเวอร์สการคูณเมทริกซ์ (Inverse matrix)

#### ให้ A เป็นเมทริกซ์มิติ n×n  ถ้า B เป็น เมทริกซ์มิติ n×n และมีสมบัติว่า

#### AB=BA=In
#### เมื่อ In เป็นเมทริกซ์เอกลักษณ์แล้วเราเรียก B ว่าเป็นเมทริกซ์อินเวอร์สของ A และเขียน B แทนด้วย A^−1

### การหา Inverse ของ Matrix ขนาด 2*2

### การหา Inverse ของ Matrix ขนาด 3*3

ในการหาอินเวอร์สของเมตริกซ์ที่มีมิติ n x n (n ≥ 3) โดยทั่วไปมี 2 วิธี

1. การหาอินเวอร์สโดยใช้เมตริกซ์ผูกผัน(Adjoint Matrix) -> นิยาม สำหรับเมตริกซ์ A ที่มีขนาด n x nถ้า cijเป็นโคแฟกเตอร์ในแถวที่ I หลักที่ j ของเมตริกซ์ A โดยที่ i,j=1,2,3,…,n แล้วเมตริกซ์ผูกผันของ A เขียนแทนด้วย adj.A = [Cof.A]t

2. การหาอินเวอร์สโดยใช้การดำเนินงานแบบแถว -> สำหรับเมตริกซ์ A ที่มีขนาด nxn การหาอินเวอร์สของเมตริกซ์ A สามารถทาได้โดยเติม เมตริกซ์เอกลักษณ์ขนาด nxn เข้ากับเมตริกซ์ A ซึ่งสามารถเขียนแทนได้ด้วยเมตริกซ์ [A: I] เมตริกซ์ ดังกล่าวเรียกว่าเป็นเมตริกซ์แต่งเติม (Augumented Matrix) จากนั้นจะใช้การดำเนินงานแบบแถวในการที่จะเปลี่ยนเมตริกซ์ [A: I] ให้เป็นเมตริกซ์ [I : B] ซึ่งจะได้ว่า B =A-1
